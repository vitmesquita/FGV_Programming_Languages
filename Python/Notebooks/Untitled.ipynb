```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-1000, 1001,1) #diz qual o tamanho do eixo
y1=2*x**2-10*x-32 #define a função em si
y0=np.zeros(2001)
```

```python
plt.plot(x,y1)
plt.plot(x,y0)
axes=plt.gca()
axes.set_xlim([-10,10])
axes.set_ylim([-50,30])
```
