```python
def retira(a,b):
    chaves1=list(a.keys())
    chaves2=list(b.keys())
    retiradas=[]
    nova_lista=a
    for k in chaves1:
        for j in chaves2:
            if k==j:
                retiradas=retiradas+[j]
            else:
                pass
    for p in retiradas:
        del nova_lista[p]
    return nova_lista
            
```
