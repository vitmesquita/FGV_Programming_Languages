```python
import time
```

```python
def counting_sort(A,k):
    tempo=time.time()
    count=[0]*(k)
    for a in A:
        count[a]+=1
    result=[]
    for b in range(len(count)):
        if count[b]!=0:
            result.extend([b])
    return (time.time()-tempo)
        
        
```

```python
def randomstring(n):
    saida=[]
    for i in range(n*10**3):
        saida.append((n*(10**3))-i)
    return saida
    
```

```python
def getcount(n,k,n_buckets):
    interval=((k-1)/n_buckets)
    balde=0
    while (n-interval)>=interval:
        balde+=1
        n+=1
    return balde
```
