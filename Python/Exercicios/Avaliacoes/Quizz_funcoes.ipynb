1 - Crie uma função que receba uma lista, um elemento, e acrescente este elemento em uma posição da lista, retornando a lista ao programa

```python
def func1(x,y,z):
    x.insert(z,y)
    return x
```

```python
func1([2,3,7,9,4],5,2)
```
