5- Explique o que são, exemplifique, e mostre algumas características de:  

+ Tipos mutáveis e tipos imutáveis  
+ Tipos "container"
+ Tipos sequenciais


```python
Tipos mutáveis são tipos que podem ser modificados e imutáveis não podem ser modificados. Tipos mutáveis possuem propriedades de modificação de seus dados como a lista que possui o ".append" e ".pop" e tipos imutáveis só possuem funções para contar seus elementos como a Tupla
```

```python
São tipos que guardam conjuntos de dados, como o Dicionario, que tem como caracteristica grupos de dados que podem ser acessados atraves de funçoes proprias do tipo.
```

```python
São tipos que guardam sequencias de elementos que seguem uma ordem denominada por um número que não temos acesso como a Lista. Se mutável, tem funçoes de somar elementos e subtrair elementos, caso contrário possui apenas funçoes de contagem e de visualização dos elementos
```
