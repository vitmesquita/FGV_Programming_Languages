```python
A=[4,7,2,9,8,5]
```

```python
def insertionsort(A):
    for i in range(len(A)-1):
        valor=A[i]
        j=i-1
        while j>=0 and A[j]>A[i]:
                A[j+1]=A[j]
                j=j-1
                A[j+1]=valor
    print(A)
```

```python
insertionsort(A)
```
