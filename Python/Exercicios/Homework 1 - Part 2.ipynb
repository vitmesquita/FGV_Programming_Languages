```python
import time
import random
```

```python
def list_generator(length):
    a = []
    for i in range(0,length):
        x = random.randint(0,2*length)
        a.append(x)
    return a
    
def insertion_sort(lista):
    for i in range(0,len(lista)):
        ordenar = lista[i]
        j = i-1
        while j>=0 and lista[j] > ordenar:
            lista[j+1] = lista[j]
            j-=1
        lista[j+1] = ordenar

def merge_sort(lista):
    if len(lista) == 1:
        return lista
    part1 = lista[0:int(len(lista)/2)]
    part2 = lista[int(len(lista)/2):len(lista)]
    return merge(merge_sort(part1),merge_sort(part2))

def merge(lista1,lista2):
    finalList = []
    while lista1 and lista2:
        if lista1[0] > lista2[0]:
            finalList.append(lista2[0])
            lista2.pop(0)
        else:
            finalList.append(lista1[0])
            lista1.pop(0)
    finalList.extend(lista1)
    finalList.extend(lista2)
    return finalList
```

### Insertion Sort

```python
#Ordem Aletória
a = 10000
while a<=100000:
    lista = list_generator(a)
    print(a)
    %time insertion_sort(lista)
    print("-------------------------------------")
    a+=10000
```

```python
#Ordem Ascendente
a = 10000
while a<=100000:
    lista = list_generator(a)
    lista.sort()
    print(a)
    %time insertion_sort(lista)
    print("-------------------------------------")
    a+=10000
```

```python
#Ordem Descendente
a = 10000
while a<=100000:
    lista = list_generator(a)
    lista.sort(reverse = True)
    print(a)
    %time insertion_sort(lista)
    print("-------------------------------------")
    a+=10000
```
