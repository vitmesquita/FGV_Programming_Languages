```python
vezes=input()
vezes=int(vezes)
listnum=[]
for a in range(vezes):
    f=input()
    listnum.append(f)
for b in listnum:
    leds=0
    for a in b:
        if a=='0' or a=='6' or a=='9':
            leds=leds+6
        if a=='2' or a=='3' or a=='5':
            leds=leds+5
        if a=='1':
            leds=leds+2
        if a=='4':
            leds=leds+4
        if a=='7':
            leds=leds+3
        if a=='8':
            leds=leds+7
    print('{} leds'.format(leds))
```

```python
lista=input()
lista=lista.split()
A=float(lista[0])
B=float(lista[1])
C=float(lista[2])
delta= B**2-4*A*C
if delta>0 or delta==0:
    if A==0:
        print('Impossivel calcular')
    else:
        r1=(-B+delta**(1/2))/(2*A) 
        r2=(-B-delta**(1/2))/(2*A)
        print('R1 = {:.5f}'.format(r1))
        print('R2 = {:.5f}'.format(r2))
else:
    print('Impossivel calcular')
```
