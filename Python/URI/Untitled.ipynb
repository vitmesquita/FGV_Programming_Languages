```python
def circulo(R):
    n=3.14159
    r=float(R)
    area=n*r**2
    print('A={:.4f}'.format(area))
    
```

```python
circulo(100.64)
```

```python
def somasim(a,b):
    soma=0
    soma=soma+a+b
    print(soma)
```

```python
somasim(0,0)
```

```python
def prod(a,b):
    prod=1
    prod=prod*a*b
    print('PROD = {}'.format(prod))
```

```python
prod(5,6)
```

```python
prod(-8,5)
```

```python
def average1(a,b):
    s=(a*3.5+b*7.5)/11
    print('MEDIA = {:.5f}'.format(s))
```
