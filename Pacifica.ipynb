```python
from estrategias.jogadores import Jogador
import math
import random

class MeuJogador(Jogador):
    def media_reputacao(self,reputacoes_dos_jogadores):
        media = sum(reputacoes_dos_jogadores)/len(reputacoes_dos_jogadores)
        return media
        
    def desvio_padrao(self,reputacoes_dos_jogadores):
        lista1=[]
        lista1= [lista1.append((x-self.media_reputacao)**2) for x in reputacoes_dos_jogadores]
        dvp=(sum(lista1)/len(lista1))**1/2 #acho o desvio padrao
        return dvp
        
    def escolha_de_cacada(self, rodada, comida_atual, reputacao_atual, m, reputacoes_dos_jogadores):
        escolhas=[]
        for k in  reputacoes_dos_jogadores:
            if comida_atual<4*len(reputacoes_dos_jogadores):
                escolhas.append('d')
            else:
                a= self.desvio_padrao* self.media_reputacao 
                b= 1-a
                if a>b:
                    escolhas.append('c')
                elif a<b:
                    escolhas.append('d')
                else:
                    e=random.randint(0,1)
                    if e==0:
                        escolhas.append('c')
                    else:
                        escolhas.append('d')
        return escolhas
```
