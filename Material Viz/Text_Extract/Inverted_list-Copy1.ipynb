```python
F1="O poeta é um fingidor"
F2="Finge tão completamente fingidor"
F3="Chega a fingir que é dor"
F4="A dor que deveras sente"
```

```python
def word_list(text):
    #fazendo o texto virar lista de palavras
    lista = text.split('\n')
    text = ' '.join(lista)
    lista = text.split(' ')
    return lista
```

```python
F1=word_list(F1)
F2=word_list(F2)
F3=word_list(F3)
F4=word_list(F4)
#transformando as frases em listas de palavras
```

```python
def idictionary(lista,n):
    #lista é o texto inserido
    #n é o numero correspondente ao texto
    d={}
    for p in range(len(lista)):
        palavra = lista[p].strip().lower()
        #a tupla vai ser feita como (numeração do texto, posição no texto)
        if palavra not in d:
            d[palavra] = []
            d[palavra].append((n,p))
        else:
            d[palavra].append((n,p))
            
    return d    
```

```python
def adictionary(lista):
    #lista de dicionarios é inserida
    defdic={}
    for k in lista:
        #concateno os valores dos dicionarios onde as chaves sao iguais
        for palavra in list(k):
            if palavra not in defdic:
                defdic[palavra]=k[palavra]
            else:
                defdic[palavra].extend(k[palavra])
    return defdic
            
```

```python
k=[idictionary(F2,2),idictionary(F1,1),idictionary(F3,3),idictionary(F4,4)]
```
