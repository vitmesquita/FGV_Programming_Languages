```python
// testando classes

#include <iostream>

using namespace std;

class Retangulo {
	int altura,largura;
  public:
	void dar_valores(int,int);
	int area() {return altura*largura;}
	int perimetro() {return 2*altura+2*largura;}
		
};

void Retangulo::dar_valores(int x, int y){
 	altura=x;
 	largura=y;
}

int main(){
	Retangulo rect1;
	rect1.dar_valores(10,10);
	cout<< "area:" << rect1.area() << endl;
	cout<< "perimetro:" << rect1.perimetro()<<endl;
	return 0
;}
```
